## 1. Distance

In all the clustering algrithms, we need define a metric that measures the differences of locations or "distances" between points in the high-dimensional space. There are many definition of this metric, but the general one is the Minkowski distance:

$$ \| X-Y  \|_p = ( \sum_{i=1}^{n} | x_i - y_i | ^{p} ) ^{1/p}$$

, where $p\ne 0$, and $X=(x_1, x_2, ..., x_n)$ and $Y=(y_1, y_2, ..., y_n)$ are two points in n-dimensional space $\mathbb{R}^{n}$.

When p=1: Manhattan distance;

p=2: Euclidean distance;

p=$+\infty$: Chebyshev distance $\| X-Y \|_{+\infty} = \max |x_i-y_i|$

p=$-\infty$: $\| X-Y \|_{-\infty} = \min |x_i-y_i|$

        
#### Domain of numbers (set) + Structure of calculating = Group or Space 

e.g. n-dimensional real numbers + linear operator = linear space on $\mathbb{R}^n$

$\mathbb{R}^n$ + $L^p$-norm = $L^p$ space (or named Lebesgue space, which is one class of Banach spaces)

$\mathbb{R}^n$ (n can be $+\infty$) + $L^2$-norm = square-integrable space (one class of Hilbert spaces)



```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

r  = 0.9
x  = r*np.cos(np.linspace(0,np.pi*2,400))
ps = [0.5, 1, 2, 4, 99]

plt.figure(1, figsize=(len(ps)*3, 2.5))
for i in range(len(ps)):
    y = x *0
    y[:len(x)/2] = (r**ps[i] - np.abs(x[:len(x)/2])**ps[i])**(1.0/ps[i])
    y[len(x)/2:] = -(r**ps[i] - np.abs(x[len(x)/2:])**ps[i])**(1.0/ps[i])
    plt.subplot(1,len(ps),i+1)
    plt.plot(x, y, 'r')
    plt.title('p='+format(ps[i]))
```

## 2. Manifold and Dimension Reduction

In the high-dimensional feature space of data (e.g. image space), the distributions of samples (e.g. faces, handwritting) ususally have lower-dimensinal structure (i.e. manifold), which are corresponding to the eignspaces (the space formed by certain eigenvectors) of the data. If there is not such a lower eignspace of the data (e.g. pure Gaussian noise), it means the signals are too weak to be identified, then it would be meanless to perform machine learning algrithms.

Examples of manifold: lines in 2-D space; lines and planes in 3-D space; m-dimensional objects in n-dimensional space (m<n)

Note the manifolds meansioned here refer to "smooth" manifolds.

** Why the concept of manifold is so important? **

Because the definition of distances above may be invalid for lower-dimensional manifold. Then, special treatments should be performed rather than the direct clustering, e.g. performing PCA firstly, and then clustering; or using special definitions on the distances.

** Three important numbers **
1. Sample number
2. Feature number (i.e. dimension number or degree of freedom)
3. Cluster number

```python
# Example of clustering of lines (1-D manifold) in 2D space
from sklearn import cluster, datasets

n = 500
data, label = datasets.make_circles(n_samples=n, factor=.5, noise=.05)

plt.plot(data[label==0,0], data[label==0,1], 'b.')
plt.plot(data[label==1,0], data[label==1,1], 'r.')
```

```python
c1 = cluster.MiniBatchKMeans(n_clusters=2)
c1.fit(data)

c2 = cluster.SpectralClustering(n_clusters=2, \
                                eigen_solver='arpack', affinity="nearest_neighbors")
c2.fit(data)

plt.figure(1, figsize=(7, 3))
plt.subplot(1,2,1)
plt.plot(data[c1.labels_==0,0], data[c1.labels_==0,1], 'b.')
plt.plot(data[c1.labels_==1,0], data[c1.labels_==1,1], 'r.')

plt.subplot(1,2,2)
plt.plot(data[c2.labels_==0,0], data[c2.labels_==0,1], 'b.')
plt.plot(data[c2.labels_==1,0], data[c2.labels_==1,1], 'r.')
```

```python
n = 500; nclass = 3; slopes=[0.5, 1, 3]; cs='rgb'
data = np.zeros([n, 2])
label = np.random.uniform(low=0, high=nclass, size=n).astype(int)

for i in range(nclass):
    r   = np.linspace(0, 1, len(label[label==i]))
    rp  = np.random.normal(loc=0.0, scale=0.02, size=len(label[label==i]))
    ang = np.random.uniform(low=0, high=2*np.pi, size=len(label[label==i]))
    data[label==i, 0] = np.sqrt(r**2/(1+slopes[i]**2)) + rp*np.cos(ang)
    data[label==i, 1] = np.sqrt(r**2/(1+1.0/slopes[i]**2)) + rp*np.sin(ang)
    plt.plot(data[label==i,0], data[label==i, 1], cs[i]+'.')
    
```

```python
c1 = cluster.MiniBatchKMeans(n_clusters=3)
c1.fit(data)

c2 = cluster.SpectralClustering(n_clusters=3, \
                                eigen_solver='arpack', affinity="nearest_neighbors")
c2.fit(data)

plt.figure(1, figsize=(9, 4))
plt.subplot(1,2,1)
for i in range(nclass):
    plt.plot(data[c1.labels_==i,0], data[c1.labels_==i,1], cs[i]+'.')

plt.subplot(1,2,2)
for i in range(nclass):
    plt.plot(data[c2.labels_==i,0], data[c2.labels_==i,1], cs[i]+'.')
```

```python
# one idea of solving this problem: projection to another space

projdata = data*0
projdata[:,0] = np.sqrt(data[:, 0]**2 + data[:, 1]**2)
projdata[:,1] = np.arctan(data[:, 1]/data[:, 0])


for i in range(nclass):
    plt.plot(projdata[label==i,0], projdata[label==i, 1], cs[i]+'.' )
```

```python
c2 = cluster.SpectralClustering(n_clusters=3, \
                                eigen_solver='arpack', affinity="nearest_neighbors")
c2.fit(projdata)

plt.figure(1, figsize=(12, 6))
plt.subplot(1,2,1)
for i in range(nclass):
    plt.plot(projdata[c2.labels_==i,0], projdata[c2.labels_==i,1], cs[i]+'.')

plt.subplot(1,2,2)
for i in range(nclass):
    plt.plot(projdata[c2.labels_==i,0]*np.cos( projdata[c2.labels_==i,1]), \
             projdata[c2.labels_==i,0]*np.sin( projdata[c2.labels_==i,1]), cs[i]+'.')
```

```python
# add a scale factor on the coordinates: "variance uniformly loaded"
fac = 10
projdata = data*0
projdata[:,0] = np.sqrt(data[:, 0]**2 + data[:, 1]**2)
projdata[:,1] = np.arctan(data[:, 1]/data[:, 0]) * fac

c2 = cluster.SpectralClustering(n_clusters=3, \
                                eigen_solver='arpack', affinity="nearest_neighbors")
c2.fit(projdata)

plt.figure(1, figsize=(12, 6))
plt.subplot(1,2,1)
for i in range(nclass):
    plt.plot(projdata[c2.labels_==i,0], projdata[c2.labels_==i,1], cs[i]+'.')

plt.subplot(1,2,2)
for i in range(nclass):
    plt.plot(projdata[c2.labels_==i,0]*np.cos( projdata[c2.labels_==i,1]/fac), \
             projdata[c2.labels_==i,0]*np.sin( projdata[c2.labels_==i,1]/fac), cs[i]+'.')
```

```python
# corvariance matrix
nx = ny = 28
mdata = np.mean(mnist.train.images, axis=0)
data = mnist.train.images - mdata
cov  = np.dot( data.T, data)/(nx*ny)

plt.figure(1, figsize=(8, 8))
plt.imshow(cov, interpolation='None')
```

```python
# eigenvalue and eigenvector of the corvariance matrix

u, s, v = np.linalg.svd(cov)

print u.shape, s.shape, v.shape
```

```python
# explained variance
ptruns = [0.90, 0.99]  # truncated percentage of explained variance
ntruns = [ np.argmin(np.abs(np.cumsum(s/np.sum(s)) - ptrun )) for ptrun in ptruns]
print [ [ntrun, np.cumsum(s/np.sum(s))[ntrun]] for ntrun in ntruns]

plt.figure(1, figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(s/np.sum(s), 'r')
for ntrun in ntruns:
    plt.plot([ntrun, ntrun], [0, max(s/np.sum(s))], 'b')

plt.subplot(1,2,2)
plt.plot(np.log(s/np.sum(s)), 'r')
for ntrun in ntruns:
    plt.plot([ntrun, ntrun], [np.log(min(s/np.sum(s))), np.log(max(s/np.sum(s))) ], 'b')
```

```python
# show the top eigenvectors

nrow = 3; ncol = 4; ntop = nrow*ncol

plt.figure(1, figsize=(3*ncol, 3*nrow))
for i in range(ntop):
    plt.subplot(nrow, ncol, i+1)
    plt.imshow(u[:,i].reshape(nx,ny), interpolation='None')
    plt.title('Top'+format(i+1)+': s='+format(s[i]/np.sum(s)*100, '.2f')+'%')
```

## 3. Clustering

### 3.1 K-Mean clustering

Given data with n samples in m-dimensional space: $X = \{x_1, x_2, ..., x_n\} \subset \mathbb{R}^{m} $, K-Means clustering algrithm classifies it into K clusters, based on the following steps:

1. **Initialization**: randomly initialize the centroids of the K clusters;

2. **Distance calculation**: calculate the distance from each sample to each centroid;

3. **Assignment**: for every sample, assign its label to the centroid with the minimum distance;

4. **Update**: for each centroid, update its position with the averaged location of the samples with the same corresponding label;

5. **Iteration**: repeat 2,3,4 until the convergence (depends on the definition of the objective function).


Objective function (distortion measure):

$$ J = \sum_{i=1}^{n} \sum_{k=1}^{K} r_{ik} \| x_i - \mu_k \|^2 $$

, where $r_{ik}=1$ if $x_i$ is assigned to cluster k, and $r_{ij}=0$ for $j \ne k$.

### 3.2. Mini Batch K-Means

The MiniBatchKMeans is a variant of the KMeans algorithm which uses mini-batches to reduce the computation time, while still attempting to optimise the same objective function. Mini-batches are subsets of the input data, randomly sampled in each training iteration. These mini-batches drastically reduce the amount of computation required to converge to a local solution. In contrast to other algorithms that reduce the convergence time of k-means, mini-batch k-means produces results that are generally only slightly worse than the standard algorithm.

See: http://scikit-learn.org/stable/auto_examples/cluster/plot_mini_batch_kmeans.html
